```python
sns.scatterplot(
    x='sepal_length',
    y='sepal_width',
    hue='species',
    data=df
)
plt.show()
```

Scatter Plot shows relationship between sepal length & width.

```python
df.hist(figsize=(10,6))
plt.show()
```

Histogram shows distribution of values.

```python
sns.boxplot(data=df)
plt.show()
```

Box Plot helps identify outliers.

Explanation: From the visualizations, we observe that Setosa species is clearly separable from others. Petal length shows strong variation among species.
